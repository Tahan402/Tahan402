#Merge/Join two dataframes

### Merge/Join two dataframes <div><center> <img src="https://pandas.pydata.org/docs/_images/08_merge_left.svg" width="1000"/> </center></div>

```python
passenger_detail = pd.DataFrame({
    'passenger_id':['P0','P2','P5','P12'],
    'gender':['Male','Male','Female','Male'],
    'age':[25,52,28,36]
})
```

```python
passenger_detail
```

```python
passenger_fare_details = pd.DataFrame({
    'passenger_id':['P0','P4','P5','P12'],
    'class':['First','Second','Third','Third'],
    'fare':[250,520,560,980]
})
```

```python
passenger_fare_details
```

```python
passenger_detail.merge(passenger_fare_details,on = ['passenger_id'],how = 'left')
```

```python
merge_df_left = passenger_detail.merge(passenger_fare_details,on = ['passenger_id'],how = 'left')
```

```python
merge_df_left
```

```python
merge_df_right = passenger_detail.merge(passenger_fare_details,on = ['passenger_id'],how = 'right')
merge_df_right
```

```python
passenger_detail.merge(passenger_fare_details,on = ['passenger_id'])
```

```python
passenger_detail.merge(passenger_fare_details,on = ['passenger_id'],how = 'outer')
```

1.   You can also fill null values based on a group
2.   For example, age for Male passengers is filled based on the average age of male passengers. Similarly, we do the same thing for female passengers.

Imputing/Replacing the missing age based on the avg. age of each gender


```python
 #we are relaoding titanic dataset so has to have missing values
titanic_data = sns.load_dataset('titanic')
```

```python
titanic_data.head(25)
```

```python
titanic_data.isnull().sum()
```

```python
titanic_data['age'] = titanic_data.groupby('sex')['age'].apply(lambda x: x.fillna(x.mean()))
titanic_data.tail()
```

NOTE: Lambda is a function where we can give instruction to do an operation (assigning a value or a mathematical operation etc) on a particular variable (e.g. age) based on the grouping of one or more other variables (e.g. sex alone or sex and class etc). Grouping can be done on multiple variables/columns


```python
titanic_data.isnull().sum()
```

Similar to above, the missing age values for males is also substituted and can be viewed using the titanic.head() command as the missing values for males are at the top rows


```python
titanic_data.head(25)
```
