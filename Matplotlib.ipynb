#Introduction to Matplotlib


Matplotlib is the "grandfather" library of data visualization with Python. It was created by John Hunter. He created it to try to replicate MatLab's (another programming language) plotting capabilities in Python. So if you happen to be familiar with matlab, matplotlib will feel natural to you.

It is an excellent 2D and 3D graphics library for generating scientific figures.

Some of the other major advantages of Matplotlib include:

*   Generally easy to get started for simple plots
*   Support for custom labels and texts
*   Great control of every element in a figure
*   High-quality output in many formats
*   Very customizable in general

Matplotlib allows you to create reproducible figures programmatically. Let's learn how to use it! Before continuing this lecture, I encourage you just to explore the official Matplotlib web page: http://matplotlib.org

We can also plot one or both axes in log scale.


```python
plt.plot(a,b, label = '4x')
plt.plot(a,c,label='$e^x$')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Plot title')

plt.yscale('log')
plt.legend()
plt.show()
```

We can also style the plots to give it visual appeal

If you're familiar with MATLAB, you'll be able to identify the styling given in the first line (the format string, which is the 3rd argument to the plot function). The format of the string is [marker][linestyle][color]

However, the styling method given in the second line is recommended (give keyword arguments rather than a format string). This way, there will be less confusion to you, and to others who read your code.


```python
plt.plot(a,b, 's-.k', label = '4x')
plt.plot(a,c, label = '$e^x$', marker = 'o', linestyle='-',linewidth=3, color = '#437293')

plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')

plt.grid()
plt.legend()
plt.show()
```

There is another way to display plots, which is referred to as the object oriented way.

Every plot consists of a "figure" that acts as a container for plots, and "axes" on which plots are drawn. (See figure for reference)


[Reference](https://towardsdatascience.com/data-visualization-for-machine-learning-and-data-science-a45178970be7)

![Image-1](https://miro.medium.com/max/700/1*b8AD_mSwFAKmizgLoIbWuA.png)
![Image-2](https://miro.medium.com/max/700/1*LCncEfpLwzktg8d5rAsvyg.png)


Plotting figures this way requires a little more code, but it offers you customizability in exchange.


```python
fig, ax=plt.subplots()

ax.plot(a,b, label = '4x')
ax.plot(a,c,label='$e^x$')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Plot title')

plt.legend()
plt.show()
```

One of the benefits of using the object oriented way of plotting is that you can create multiple plots within a figure.


```python
fig , ax= plt.subplots(1,2,figsize=(15,6))

ax[0].plot(a,b)
ax[0].set_xlabel('X')
ax[0].set_ylabel('Y')
ax[0].set_title('$y=4x$')

ax[1].plot(a,c)
ax[1].set_xlabel('X')
ax[1].set_ylabel('Y')
ax[1].set_title('$y=e^x$')

plt.show()
```

# Downloading and using a dataset


Plotting data from a dataset gives you a better understanding about how different plots can convey information in their own ways. 

Let's use the Iris Dataset for plotting.


```python
import requests

def download_file(url, dest):
    try:
        response = requests.get(url)
        with open(dest,'wb') as f:
            f.write(response.content)
        print("File Downloaded.")
    except:
        print("Error: Download Failed.")
        
dataset_url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
dataset_dest = 'iris.csv'

download_file(dataset_url, dataset_dest)
```

```python
import seaborn as sns
 
iris = sns.load_dataset('iris')
```

The Iris Dataset is one of the most popular datasets found in pattern recognition literature. It is an old dataset that was published in July 1988, curated by R. A. Fisher.


The dataset consists of 3 classes of 50 instances each (total 150 data samples). Each sample has 4 numerical features/attributes, namely sepal length, sepal width, petal length and petal width. In a classification problem, these features are used to determine which class a test sample belongs to.

You can read more about the dataset in the link:
https://archive.ics.uci.edu/ml/datasets/Iris
